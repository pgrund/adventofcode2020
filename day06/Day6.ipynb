```python
import pandas as pd
import numpy as np

answers_file = open('input', mode='r')

data = [group.split('\n') for group in answers_file.read().split('\n\n')]
data
```

```python
dummy = """abc

a
b
c

ab
ac

a
a
a
a

b"""


data = [group.split('\n') for group in dummy.split('\n\n')]
```

## Part 1

```python

group_count=[]
for group in data:
    result={}
    for person in group:
         for answer in person:
            result[answer]=True
    group_count.append(len(result))

print(sum(group_count))
group_count
```

## Part 2

```python


group_count=[]
for group in data:
    result={}
    for person in group:
         for answer in person:
            result[answer]= result[answer]+1 if answer in result else 1
    
    filtered_dict = {k:v for (k,v) in result.items() if v==len(group)}
    group_count.append(len(filtered_dict))

print(sum(group_count))
group_count
```
