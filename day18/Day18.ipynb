```python
data = [[l  for l in equation if l != " "] for equation in homework.splitlines()]
["".join(d) for d in data]
```

## Part 1

```python
ops = {
    '+': lambda x,y: x+y,
    '-': lambda x,y: x-y,
    '*': lambda x,y: x*y,
    '/': lambda x,y:x/y
}
result=[]
for d in data:
    #print("".join(d))
    stack = []
    funcs = []
    for n in d:
        if n in ops:
            funcs.append(ops[n])
        elif n == '(':
            funcs.append(None)
        elif n == ')':
            func = funcs.pop() if len(funcs)>0 else None
            b = stack.pop()
            stack.append(func(stack.pop(),b)if func != None else b)
        else: 
            func = funcs.pop() if len(funcs)>0 else None
            if func != None:
                stack.append(func(stack.pop(), int(n)))
            else:
                stack.append(int(n))
    
    result.append(stack.pop())
    
sum(result)
```
