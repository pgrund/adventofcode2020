## Part 1

```python
facing='E'
directions = { 
    'N':(0,-1),
    'S':(0,1),
    'E':(1,0),
    'W':(-1,0),
    'F': (0,0),
    'L':(0,0),
    'R':(0,0)
}
facing_switch = {
    90: {
        'R': {
            'N':'E',
            'E':'S',
            'S':'W',
            'W':'N'
        },
        'L':{
            'N':'W',
            'W':'S',
            'S':'E',
            'E':'N'
        }
    },
    180:{
        'R': {
            'N':'S',
            'E':'W',
            'S':'N',
            'W':'E'
        },
        'L':{
            'N':'S',
            'W':'E',
            'S':'N',
            'E':'W'
        }},
    270:{       
        'R': {
            'N':'W',
            'E':'N',
            'S':'E',
            'W':'S'
        },
        'L':{
            'N':'E',
            'W':'N',
            'S':'W',
            'E':'S'
        }},
    360:{
        'L':{
            'N':'N',
            'W':'W',
            'S':'S',
            'E':'E'
        },
        'R':{
            'N':'N',
            'W':'W',
            'S':'S',
            'E':'E'
        }
    }
}
start=[0,0]
def apply_rules_p1(pos, inst):
    global directions,facing
    [action,value]=[inst[:1],int(inst[1:])]
    if action in [ 'L', 'R']:
        facing=facing_switch[value][action][facing]
        action=(0,0)
    elif action == 'F':
        action=directions[facing]
    else :
        action = directions[action]
    return np.add(pos,np.multiply(action,value))
    

sum(reduce(apply_rules_p1,instructions,start))
```

## Part 2

```python
waypoint=[10,-1]
ship=[0,0]
facing_switch={
    'R': {
        90: lambda pos: [-pos[1],pos[0]], # passt
        180: lambda pos : [-pos[0],-pos[1]],
        270: lambda pos : [pos[1], -pos[0]],
        360: lambda pos: pos
    },
    'L': {
        90: lambda pos: [pos[1],-pos[0]],
        180: lambda pos : [-pos[0],-pos[1]],
        270: lambda pos : [-pos[1], pos[0]],
        360: lambda pos: pos
    }
}

def apply_rules_p2(prev, inst):

    [pos, rel_way] = prev
    global directions,facing
    [action,value]=[inst[:1],int(inst[1:])]
    if action=='F':
        # only action for ship (implicit waypoint)
        pos =np.add(pos, np.multiply(rel_way,value))
    else :
        # only action on waypoint
        if action in ['L', 'R']:
            func =facing_switch[action][value]
            rel_way=func(rel_way)
        else:
            action=directions[action]
            rel_way = np.add(rel_way,np.multiply(action,value))
        
    #print(inst,prev,'->',[pos,rel_way])
    return [pos,rel_way]


sum(np.abs(reduce(apply_rules_p2,instructions,[ship,waypoint]))[0])

# ?? > 46081 
```
