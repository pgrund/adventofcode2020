```python
#slope={'right':3, 'down':1}
class Slope:
    """Class to implement an iterator
    of a slope, going right, down, ..."""

    def __init__(self, start=[0,0], inc=[0,0], max=[0,0]):
        self.max = max
        self.start = start
        self.inc = inc

    def __iter__(self):
        self.n = self.start
        return self

    def __next__(self):
        if (self.n[0] +self.inc[0]) <= self.max[0]:
            self.n = [self.n[0]+self.inc[0], (self.n[1]+self.inc[1]) % (self.max[1]+1)]
            return self.n
        else:
            raise StopIteration
            
df = trees
s = [row for row in iter(Slope(inc=[1,3],max=[df.row.max(), df.column.max()]))]
#print(s)

t = pd.DataFrame({ 'row':[ x for [x,y] in s], 'column':[ y for [x,y] in s]})
        
result = df.merge(t)
print(result.value.value_counts())
result
```
