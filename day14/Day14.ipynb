## Part 1

<pre>
value:  000000000000000000000000000000001011  (decimal 11)
mask:   XXXXXXXXXXXXXXXXXXXXXXXXXXXXX1XXXX0X
result: 000000000000000000000000000001001001  (decimal 73)
</pre>

```python
mask = 'X'*36
memory={}

def apply_mask_part1(mask, number):
    bin_str_r = list(format(number,'b')[::-1].ljust(36,'0'))
    for (idx, letter) in [(idx, letter) for idx,letter in enumerate(mask[::-1]) if letter !="X"]:
        bin_str_r[idx]= letter

    return int("".join(bin_str_r[::-1]),2)

def apply_rules_part1(command):
    global mask, memory
    [action, address, value] = command
    if action == 'mask':
        mask = value
        #print('set mask to ' + mask)
    else:
        # write to mem
        masked_value = apply_mask_part1(mask, int(value))
        #print("was {}: [{}]=> {}".format(value, address,masked_value))
        memory[address] =  int(masked_value)

for d in data:
    apply_rules_part1(d)
    
print("result:",sum(memory.values()))
#memory
```

## Part 2

```python
mask = 'X'*36
memory={}

def apply_mask_part2(mask, number):
    masked_addresses = []
    bin_str_r = list(format(number,'b')[::-1].ljust(36,'0'))
    for (idx, letter) in [(idx, letter) for idx,letter in enumerate(mask[::-1]) ]:
        if letter != '0':
            bin_str_r[idx] = letter
    floating = [idx for idx,val in enumerate(bin_str_r) if val == 'X']
    if len(floating) == 0:
        masked_addresses.append(int("".join(bin_str_r[::-1]),2))
    else :        
        for comb in  [p for p in product('01', repeat=len(floating))]:
            n = np.copy(bin_str_r)
            for comb_idx,bin_idx in enumerate(floating):
                n[bin_idx]=comb[comb_idx]
           # print("".join(n))
            masked_addresses.append(int("".join(n[::-1]),2))
    
    return sorted(masked_addresses)

def apply_rules_part2(command):
    global mask, memory
    [action, address, value] = command
    if action == 'mask':
        mask = value
        #print('set mask to ' + mask)
    else:
        # write to mem
        masked_addresses = apply_mask_part2(mask, address)
        #print("was [{}]: [{}]=> {}".format(address, masked_addresses,value))
        for add in masked_addresses:
            memory[add] =  int(value)

for d in data:
    apply_rules_part2(d)
    
print("result:",sum(memory.values()))
#memory
```
