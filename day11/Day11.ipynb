```python
import pandas as pd
import numpy as np
```

```python
seats_file = open('input', mode='r')
data = seats_file.read().splitlines()
data
```

```python
dummy = """L.LL.LL.LL
LLLLLLL.LL
L.L.L..L..
LLLL.LL.LL
L.LL.LL.LL
L.LLLLL.LL
..L.L.....
LLLLLLLLLL
L.LLLLLL.L
L.LLLLL.LL"""
data = dummy.splitlines()
```

```python
dummy_small = """L.LLL
LLLL
L.L.
LLLL
L.LL
L.LL
..L.
LLLL
L.LL
L.LL"""
data = dummy_small.splitlines()
```

```python

df = pd.DataFrame([(int(x),int(y),letter) for (x,line) in enumerate(data) for (y,letter) in enumerate(line)], columns=['row', 'column','seat'])
width=df.column.max()+1
height=df.row.max()+1

def neighbor_index(idx):
    n_idx = [
        idx-width-1,
        idx-1,    
        idx+width-1,
        idx-width,        
        idx+width,
        idx-width+1,
        idx+1,
        idx+width+1
    ] 
    if (idx%width == 0): 
        # left overflow
        n_idx = n_idx[3:]        
    elif ((idx % width +1) == width):
        # right overflow
        n_idx = n_idx[:-3]
        
    return [n for n in n_idx if (n >=0 & n< height)]

df['empty']=(df['seat']=='L').astype(int)
df['occupied']=(df['seat']=='#').astype(int)
df['n_idx']=df.index.map(neighbor_index)
df['neighbors_mask']=df.index.map(neighbor_index).map(lambda x: [n in x for n in range(df.shape[0])])

df['neighbors']=0

df1 = df.copy()
df1
```

```python
def neighbors(row):
    row['neighbors'] = df1[row['neighbors_mask']]['occupied'].sum()
    return row


for r in range(80,100):
    print("Round",r)
    old = df1.copy()
    
    df1.loc[(df1['empty']==1) & (df1['neighbors']==0),'seat']='#'
    df1.loc[(df1['occupied']==1) & (df1['neighbors']>=4),'seat']='L'    
    df1['empty']=(df1['seat']=='L').astype(int)
    df1['occupied']=(df1['seat']=='#').astype(int)

    df1=df1.apply(neighbors,axis=1)
    if df1.equals(old):
        print("stopped in round {}, occupied seats: {}".format(r, df1.occupied.sum()))
        break

print('done')
df1
```

```python
df1.to_csv('day11-part1-79.csv',encoding='utf-8')
```
