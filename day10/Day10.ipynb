## Part 1

```python
d1= df.copy().append(pd.DataFrame([0,(df.jolts.max()+3)],columns=['jolts'])).sort_values(by='jolts').reset_index().drop(columns=['index'])
d1['diff']=(d1['jolts']-d1.shift(1)['jolts'])
diffs = d1['diff'].value_counts()
print('product:', np.prod(diffs.values))
diffs
```

## Part 2

```python
def calc_next(row):
    l = [int(l) for l in (d1[row.name+1:row.name+4].jolts.values) if l <= (row.jolts +3)]
    row['next']=l
    row['a']=len(l)
    return row

d1 = d1.apply(calc_next, axis=1)
d1
```
